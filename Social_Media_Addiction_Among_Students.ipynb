DATA EXPLORATION AND CLEANING

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.dates as mdates
df=pd.read_csv('sample_data/Students Social Media Addiction.csv')
df.head()
df.info()
df.isnull().sum()
df.describe()
df.columns
```

```python
df.columns=df.columns.str.strip()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(data=df,x='Most_Used_Platform',y='Avg_Daily_Usage_Hours',palette='viridis',ci=None)
plt.xlabel('Most_Used_Platform')
plt.ylabel('Avg_Daily_Usage_Hours')
plt.title('Average Daily Usage Hours By Most Used Platforms')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(20,6))
sns.barplot(data=df,x='Country',y='Avg_Daily_Usage_Hours',palette='viridis',ci=None)
plt.xlabel('Country')
plt.ylabel('Avg_Daily_Usage_Hours')
plt.title('Average Daiy Usage Hours By country')
plt.xticks(rotation=90)
plt.show()
```

```python
top_countries=df['Country'].value_counts().head(10).index
sns.barplot(x='Country',y='Avg_Daily_Usage_Hours',ci=None,data=df[df['Country'].isin(top_countries)])
plt.title('Average Daily Usage Hours By top 10 Countries')
plt.xlabel('top 10 Countries')
plt.ylabel('Avg_Daily_Usage_Hours')
plt.xticks(rotation=90)
plt.show()
```

RELATIONSHIP BETWEEN USAGE HOURS AND ADDICTION SCORE

```python
plt.figure(figsize=(12,6))
sns.scatterplot(data=df,x='Avg_Daily_Usage_Hours',y='Addicted_Score',hue='Gender')
plt.xlabel('Avg_Daily_Usage_Hours')
plt.ylabel('Addicted_Score')
plt.title('Usage Hours vs Addicted Score')
plt.show()
```

CORRELATION MATRIX

```python
corr_matrix=df[['Avg_Daily_Usage_Hours','Addicted_Score','Mental_Health_Score','Sleep_Hours_Per_Night','Conflicts_Over_Social_Media']].corr()
sns.heatmap(corr_matrix,annot=True,cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

ACADEMIC IMPACT

```python
#Academic performance affected by social media
plt.figure(figsize=(12,6))
sns.countplot(data=df,x='Affects_Academic_Performance')
plt.title('Does social media affect academic performance?')
plt.show()

```

```python
#compare between those affected or not affected
plt.figure(figsize=(12,6))
sns.barplot(data=df,x='Affects_Academic_Performance',y='Avg_Daily_Usage_Hours')
plt.title('Usage Hours: Academic performance affected vs not affected')
plt.xlabel('Affects_Acedemis_Performance')
plt.ylabel('Avg_Daily_Usage_Hours')
plt.show()
```
